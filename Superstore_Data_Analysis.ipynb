```python
# 6. Convert the Order Date and Ship Date columns to datetime format. 

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
print(df)
```

```python
# 7. Drop the Row ID from dataset

df.drop(columns="Row ID")
```

```python
8. # Rename the Country/Region column to Country. 
df.rename(columns = {"Country/Region":"Country"})
```

```python
# 9. Check for duplicate rows in the dataset. 

print ( "Duplicated Rows: ",df.duplicated().sum())
```

```python
# 10. Remove any duplicate rows from the dataset. 

df.drop_duplicates()
```

```python
# 11. Remove $ symbol from “sales” and other special characters like”!@#$&” in the data.

df["Sales"]=df["Sales"].replace('[!@#$&]','',regex = True).astype(float)
print (df)
```

```python
df.dtypes
```

```python
# ---------------------------- DATA PREPROCESSING ----------------------------------------------
```

```python
# 12. Fill missing values in the columns with the appropriate method. 

# Fill missing values in different columns
df["Customer Name"] = df["Customer Name"].ffill()
df["City"] = df["City"].ffill()
df["State"] = df["State"].ffill()
df["Postal Code"] = df["Postal Code"].ffill()

# Fill numeric columns with mean or median
df["Profit"] = df["Profit"].fillna(df["Profit"].mean())      # Mean for Profit
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())         # Mean for Sales
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())  # Median for Quantity

# Fill categorical column with mode
df["Segment"] = df["Segment"].fillna(df["Segment"].mode()[0])

print (df.isnull().sum())
```

```python
# 13.  Replace missing values in the Discount column with 0. 
df["Discount"]=df["Discount"].fillna(0)
print ("Missing values in Discount:",df["Discount"].isnull().sum())
```

```python
# 14. Create a new column Profit Margin which is calculated as Profit / Sales.
df["Profit Margin"] = df["Profit"]/df["Sales"]
print("Profit Margin Created. \n ",df["Profit Margin"].head(10))
```

```python
# 15. Create a new column Order Year by extracting the year from the Order Date. 

df["Order Year"]=df["Order Date"].dt.year
print(" Order Year created. \n",df["Order Year"])
```

```python
# 16.  Create a new column Order Month by extracting the month from the Order Date. 

df["Order Month"]=df["Order Date"].dt.month
print(" Order Month created. \n",df["Order Month"])
```

```python
# 17. Create a new column Order Day by extracting the day from the Order Date.

df["Order Day"]=df["Order Date"].dt.day
print(" Order Day created. \n",df["Order Day"])
```

```python
# 18.  Create a new column Ship Year by extracting the year from the Ship Date.

df["Ship Year"]=df["Ship Date"].dt.year
print("Ship Year created. \n",df["Ship Year"])
```

```python
# 19. Create a new column Ship Month by extracting the month from the Ship Date.

df["Ship Month"]=df["Ship Date"].dt.month
print(" Ship Month created. \n",df["Ship Month"])
```

```python
# 20. Create a new column Ship Day by extracting the day from the Ship Date. 

df["Ship Day"]=df["Ship Date"].dt.day
print("Ship Day created. \n",df["Ship Day"])
```

```python
# 21. Convert the Category column to lowercase. 

df["Category"] = df["Category"].str.lower()
print("Category column converted to lowercase. \n",df["Category"].unique())
```

```python
# -------------------------------------- FEATURE ENGINEERING --------------------------------------
```

```python
# 22. Create a new column Total Cost using sales and profit.

df["Total Cost"]= df["Sales"]-df["Profit"]
print (" 'Total Cost' Column created.")
```

```python
# 23.  Create a new column Discount Amount which is calculated using Sales and Discount.

df["Discount Amount"] = df["Sales"]*df["Discount"]
print(" 'Discount Amount' Column created.")
```

```python
# 24.  Create a new column Net Sales which is calculated using Sales and Discount. 

df["Net Sales"] = df["Sales"] - df["Discount Amount"]
print (" 'Net Sales' Column created.")
```

```python
# 25. Create a new column “Order_to_Ship_Days” using Ship Date and Order Date. 

df["Order_to_Ship_Days"] = df["Ship Date"] - df["Order Date"]
print(df["Order_to_Ship_Days"].dt.days)
```

```python
# 26. Create a new column High Profit which is True if Profit is greater than threshold value, otherwise False.

df["High Profit"] = np.where(df["Profit"]>50,True,False)            #Threshold Value = 50
print("'High Profit' Column created. \n",df["High Profit"])
```

```python
# 27. Create a new column Sales Category which categorizes Sales into 'Low', 'Medium', and 'High' based on quantiles.

df["Sales Category"] = pd.qcut(df["Sales"], q=3, labels=['Low', 'Medium', 'High'])
print("New categorical column 'Sales Category' created.\n",df["Sales Category"])
```

```python
# 28. Create a new column Profit Category which categorizes Profit into 'Loss', 'Break Even', and 'Profit' based on the value of Profit.

df["Profit Category"] = np.where(df["Profit"] < 0, 'Loss', 
                        np.where(df["Profit"] == 0, 'Break Even', 'Profit'))
print(" New categorical column 'Profit Category' created. \n",df["Profit Category"])
```

```python
# 29. Create a new column Ship Mode Category which categorizes Ship Mode into 'Fast', 'Medium', and 'Slow' based on the “Order_to_Ship_Days” .

conditions = [df["Order_to_Ship_Days"].dt.days < 2, df["Order_to_Ship_Days"].dt.days < 5]
choices = ['Fast', 'Medium']
df["Ship Mode Category"] = np.select(conditions, choices, default='Slow')
print(" 'Ship Mode Category' Column created.\n",df[['Order_to_Ship_Days', 'Ship Mode Category']].tail(10))
```

```python
# 30. Create a new column Customer Segment which categorizes Segment into 'Individual', 'Corporate', and 'Home Office'.

segment_mapping = {'Consumer': 'Individual', 'Corporate': 'Corporate', 'Home Office': 'Home Office'}
df["Customer Segment"] = df["Segment"].map(segment_mapping)
print(" New column 'Customer Segment' created using mapping. \n",df["Customer Segment"])
```
